```python
# Importing the pipeline function from the transformers library
from transformers import pipeline
```

```python
# Creating a Text2TextGenerationPipeline for language translation
pipe = pipeline(task='text2text-generation', model='facebook/m2m100_418M')
```

```python
# Converting
pipe("Hello Teacher, I love class!", forced_bos_token_id=pipe.tokenizer.get_lang_id(lang='fr'))
```

```python
pipe("Hello Teacher, I love class!", forced_bos_token_id=pipe.tokenizer.get_lang_id(lang='hi'))
```

```python
pipe("Hello Teacher, I love class!", forced_bos_token_id=pipe.tokenizer.get_lang_id(lang='es'))
```
